## Figure 5.22

This figure illustrates evaluation of the posterior. This is _the_ reason why Bayesian inference was not popular before the availability of computing resources. The evaluation is done using Metropolis Hastings Monte Carlo Markov Chain (M-H MCMC) algorithm. Details of the algorithm are given in the textbook.

The easy bit is availability of codes to do this evaluation for us. This example uses `pymc` package. Nowadays, the `pymc3` package apparently gives more options. There are a number of available packages in Python to implement the Monte Carlo Markov Chain routine. 

The code is given at [Fig 5.22](http://www.astroml.org/book_figures/chapter5/fig_cauchy_mcmc.html)

```python
run fig_cauchy_mcmc.py
```

This code is given in [Fig 5.20](http://www.astroml.org/book_figures/chapter5/fig_hist_binsize.html) and shows _Knuth's method_
