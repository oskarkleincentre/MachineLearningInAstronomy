I have tried to fit the data with a GMM with 6 components and with an aproximate physics based model. The PBM is a function with 2 free gaussians (one for the noise, one for the 1PE peak), 2 gaussians dependent on the 1PE parameters, and 2 underlying exponential curves<br />
